```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
text_df = pd.read_csv("../../data/processed/text_df.csv", index_col=0)
tags_df = pd.read_csv("../../data/processed/tags_df.csv", index_col=0)
target_df = pd.read_csv("../../data/processed/target_df.csv", index_col=0)
```

```python
display(text_df.head())
display(tags_df.head())
display(target_df.head())
```

#### Разделим датасет на обучающую и валидационную выборки следующим образом: первые 90% данных пойдут на обучение, на оставшихся 10% будем сравнивать решения

```python
text_train_df = text_df.loc[0:int(len(text_df)*0.9)]
text_val_df = text_df.loc[int(len(text_df)*0.9) + 1:]

display(text_train_df)
display(text_val_df)

text_train_df.to_csv("../../data/processed/train/text_train_df.csv")
text_val_df.to_csv("../../data/processed/val/text_val_df.csv")
```

```python
tags_train_df = tags_df.loc[0:int(len(tags_df)*0.9)]
tags_val_df = tags_df.loc[int(len(tags_df)*0.9)+1:]

display(tags_train_df)
display(tags_val_df)

tags_train_df.to_csv("../../data/processed/train/tags_train_df.csv")
tags_val_df.to_csv("../../data/processed/val/tags_val_df.csv")
```

```python
target_train_df = target_df.loc[0:int(len(target_df)*0.9)]
target_val_df = target_df.loc[int(len(target_df)*0.9)+1:]

display(target_train_df)
display(target_val_df)

target_train_df.to_csv("../../data/processed/train/target_train_df.csv")
target_val_df.to_csv("../../data/processed/val/target_val_df.csv")
```
